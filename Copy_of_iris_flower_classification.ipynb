```python
# Assuming 'f' is a pandas DataFrame containing the data
df.plot(kind="bar", x="SepalLengthCm", y="SepalWidthCm")

# Adding labels and title
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Width vs. Sepal Length")

# Display the plot
plt.show()
```

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=df)

# Adding labels and title
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.title("Distribution of Petal Length for Different Species")

# Display the plot
plt.show()
```

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=df)

# Adding labels and title
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.title("Distribution of Sepal Length for Different Species")

# Display the plot
plt.show()
```

```python
# Create the FacetGrid with KDE plot
sns.FacetGrid(df, hue="Species", height=6) \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()

# Adding labels and title
plt.xlabel("Petal Length (cm)")
plt.ylabel("Density")
plt.title("Distribution of Petal Length for Different Species")

# Display the plot
plt.show()
```

```python
# Drop the "Id" column if it exists in the DataFrame
# Assuming 'iris' is your DataFrame containing the data
iris_cleaned = df.drop("Id", axis=1)

# Create the pair plot with 'hue'
sns.pairplot(iris_cleaned, hue="Species", height=6)

# Display the plot
plt.show()
```
