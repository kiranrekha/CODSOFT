```python

import matplotlib.pyplot as plt

# Assuming `df` contains the DataFrame with the 'Survived' column
# For example:
data = {
     'PassengerId': [892,893,894,895],
    'Name': ['Mr.James', 'Mrs.James', 'Mr.Thomas', 'Mr.Albert'],
     'Age': [34.5,47,62,27],
     'Survived':[0,1,0,0]
}

df = pd.DataFrame(data)
df["Survived"].value_counts().plot(kind="bar")

# Adding labels and title to the plot
plt.xlabel("Survived")
plt.ylabel("Count")
plt.title("Survived Passengers Count")

# Display the plot
plt.show()
```

```python

import matplotlib.pyplot as plt

# Assuming `df` contains the DataFrame with the 'Sex' column
# For example:
data = {
    'PassengerId': [892,893,894,895],
    'Name': ['Mr.James', 'Mrs.James', 'Mr.Thomas', 'Mr.Albert'],
    'Age': [34.5,47,62,27],
    'Sex': ['male', 'female', 'male', 'male']
}

df = pd.DataFrame(data)

# Plotting the pie chart of 'Sex' column counts
df["Sex"].value_counts().plot(kind="pie", autopct="%.2f%%")

# Adding title to the plot
plt.title("Passenger Gender Distribution")

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Display the plot
plt.show()
```

```python

import matplotlib.pyplot as plt

# Assuming `df` contains the DataFrame with numerical columns
# For example:
data = {
       'PassengerId': [892,893,894,895],
     'Age': [34.5,47,62,27],
     'Pclass':[3,3,2,3]
}

df = pd.DataFrame(data)

# Plotting histograms for all numerical columns in `df`
df.hist(figsize=(15, 15))

# Adding a title for the overall plot
plt.suptitle("Histograms for Numeric Columns", fontsize=16)

# Display the plot
plt.show()
```
